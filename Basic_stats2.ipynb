## Imports and Data Setup

```python
import numpy as np
import scipy.stats as stats

data = np.array([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29])

n = len(data)
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1) 

print(f"--- Sample Statistics ---")
print(f"Sample Size (n): {n}")
print(f"Sample Mean (x_bar): {sample_mean:.4f} million characters")
print(f"Sample Standard Deviation (s): {sample_std:.4f} million characters")
```

## Task A – 99% Confidence Interval (Population Standard Deviation Unknown)

```python
print("=== TASK A: 99% Confidence Interval (t-distribution) ===")

confidence_level = 0.99
df = n - 1  # Degrees of freedom = 14


t_critical = stats.t.ppf((1 + confidence_level) / 2, df)

# 3. Calculate Standard Error of the Mean (SEM)
sem_t = sample_std / np.sqrt(n)

# 4. Calculate Margin of Error (ME)
margin_of_error_t = t_critical * sem_t

# 5. Calculate Confidence Interval bounds
ci_lower_t = sample_mean - margin_of_error_t
ci_upper_t = sample_mean + margin_of_error_t

print(f"Degrees of Freedom: {df}")
print(f"Critical t-value: {t_critical:.4f}")
print(f"Standard Error: {sem_t:.4f}")
print(f"Margin of Error: {margin_of_error_t:.4f}")
print(f"99% Confidence Interval: ({ci_lower_t:.4f}, {ci_upper_t:.4f})")
print("\n--- Explanation ---")
print(f"We are 99% confident that the true population mean durability of the print-heads ")
print(f"lies between {ci_lower_t:.4f} and {ci_upper_t:.4f} million characters.")
```

## Task B – 99% Confidence Interval (Population Standard Deviation Known)

```python
print("=== TASK B: 99% Confidence Interval (z-distribution) ===")

pop_std = 0.2
confidence_level = 0.99

# 2. Find the critical z-value
# For a 99% CI, alpha = 0.01, so we look at the 0.995 percentile
z_critical = stats.norm.ppf((1 + confidence_level) / 2)

# 3. Calculate Standard Error using population standard deviation
sem_z = pop_std / np.sqrt(n)

# 4. Calculate Margin of Error
margin_of_error_z = z_critical * sem_z

# 5. Calculate Confidence Interval bounds
ci_lower_z = sample_mean - margin_of_error_z
ci_upper_z = sample_mean + margin_of_error_z

# Display results
print(f"Known Population Standard Deviation (sigma): {pop_std}")
print(f"Critical z-value: {z_critical:.4f}")
print(f"Standard Error: {sem_z:.4f}")
print(f"Margin of Error: {margin_of_error_z:.4f}")
print(f"99% Confidence Interval: ({ci_lower_z:.4f}, {ci_upper_z:.4f})")
print("\n--- Explanation ---")
print(f"With a known population standard deviation of 0.2, we are 99% confident that ")
print(f"the true population mean durability lies between {ci_lower_z:.4f} and {ci_upper_z:.4f} million characters.")
```
