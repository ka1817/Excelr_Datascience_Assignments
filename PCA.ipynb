```python
import math

df = pd.read_csv('wine.csv')

numeric_cols = df.select_dtypes(include=['float']).columns

cols_per_row = 3
n_rows = math.ceil(len(numeric_cols) / cols_per_row)

plt.figure(figsize=(15, n_rows * 4))

for i, column in enumerate(numeric_cols):
    plt.subplot(n_rows, cols_per_row, i + 1)
    sns.boxplot(y=df[column], color='skyblue')
    plt.title(f'Boxplot of {column}')

plt.tight_layout()
plt.show()
```

Significant Outliers: Several features—most notably Magnesium, Color Intensity, and Malic Acid—show distinct data points far above the top whisker. This means these columns have "extreme" high values that differ significantly from the average wine sample.

Massive Scale Differences: The features are on completely different scales. For example, Proline values reach up to 1500, while Nonflavanoids are all below 1.0. This is a huge hint that you must use Feature Scaling (like StandardScaler) before running any Machine Learning models.

Feature Variability: Proline and Alcalinity of Ash have very large boxes (Interquartile Ranges), showing that these values vary wildly across different wines. In contrast, Ash and Phenols have much tighter boxes, indicating more consistency.

Skewness: Features like Malic Acid, Magnesium, and Color Intensity have a median line (the line inside the box) that is closer to the bottom of the box, with a long whisker at the top. This confirms the data is right-skewed.

Classification Potential: The "Type" column shows three flat lines because it's a categorical label (1, 2, or 3). The fact that the other chemical features show such distinct spreads suggests they will be very effective at helping a model predict which "Type" a wine belongs to.

Independent Features: Ash and Magnesium show very low correlation scores with most other variables (often near 0.00 to 0.20). This means these chemical properties vary independently and aren't strongly tied to the presence of other minerals or compounds in the wine.

Hue and Color Inverse Relationship: There is a significant negative correlation (-0.52) between Hue and Color Intensity. This makes sense physically; as color becomes more intense and deep, the "hue" (often representing the shade or brightness) tends to shift or decrease in value.

## KMeans Before PCA

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
X = df.drop('Type', axis=1)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

wcss = []  # Within-Cluster Sum of Square
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 5))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method to Find Optimal Clusters')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS (Inertia)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

```

```python
kmeans_orig = KMeans(n_clusters=3, init='k-means++', random_state=42)
labels_orig = kmeans_orig.fit_predict(X_scaled)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
kmeans_pca = KMeans(n_clusters=3, init='k-means++', random_state=42)
labels_pca = kmeans_pca.fit_predict(X_pca)

plot_df = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
plot_df['Clustering_All_Features'] = labels_orig
plot_df['Clustering_PCA_Only'] = labels_pca

plt.figure(figsize=(16, 6))

plt.subplot(1, 2, 1)
sns.scatterplot(data=plot_df, x='PC1', y='PC2', hue='Clustering_All_Features', palette='viridis', s=80)
plt.title('K-Means: Clustered on All 13 Features\n(Visualized with PCA)')

plt.subplot(1, 2, 2)
sns.scatterplot(data=plot_df, x='PC1', y='PC2', hue='Clustering_PCA_Only', palette='magma', s=80)
plt.title('K-Means: Clustered on 2 Principal Components\n(Clustered & Visualized with PCA)')

plt.tight_layout()
plt.show()
```

Evaluation and Interpreting Explained variance

```python
import numpy as np
from sklearn.metrics import silhouette_score, davies_bouldin_score
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca_full = PCA()
pca_full.fit(X_scaled)

exp_var = pca_full.explained_variance_ratio_
cum_var = np.cumsum(exp_var)

plt.figure(figsize=(10, 5))
plt.bar(range(1, 14), exp_var, alpha=0.5, align='center', label='Individual Variance')
plt.step(range(1, 14), cum_var, where='mid', label='Cumulative Variance')
plt.axhline(y=0.90, color='r', linestyle='--', label='90% Threshold')
plt.ylabel('Explained Variance Ratio')
plt.xlabel('Principal Component Index')
plt.title('Scree Plot & Cumulative Variance')
plt.legend(loc='best')
plt.savefig('pca_scree_plot.png')

pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_scaled)

kmeans_orig = KMeans(n_clusters=3, init='k-means++', random_state=42)
clusters_orig = kmeans_orig.fit_predict(X_scaled)

sil_orig = silhouette_score(X_scaled, clusters_orig)
db_orig = davies_bouldin_score(X_scaled, clusters_orig)

kmeans_pca = KMeans(n_clusters=3, init='k-means++', random_state=42)
clusters_pca = kmeans_pca.fit_predict(X_pca)

sil_pca = silhouette_score(X_pca, clusters_pca)
db_pca = davies_bouldin_score(X_pca, clusters_pca)

print(f"--- Performance Metrics ---")
print(f"Original Data -> Silhouette: {sil_orig:.4f}, DB Index: {db_orig:.4f}")
print(f"PCA Data (3PC)-> Silhouette: {sil_pca:.4f}, DB Index: {db_pca:.4f}")

y = df['Type'] 

plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=clusters_orig, palette='viridis')
plt.title('Clustered on Original Data')

plt.subplot(1, 3, 2)
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=clusters_pca, palette='magma')
plt.title('Clustered on PCA Components')

plt.subplot(1, 3, 3)
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=y, palette='tab10')
plt.title('Original Wine Types (Ground Truth)')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# 1. Get the numerical values
variance = pca.explained_variance_
variance_ratio = pca.explained_variance_ratio_
cumulative_variance = variance_ratio.cumsum()

pca_summary = pd.DataFrame({
    'PC': [f'PC{i+1}' for i in range(len(variance_ratio))],
    'Eigenvalue (Variance)': variance,
    'Explained Variance Ratio': variance_ratio,
    'Cumulative Variance': cumulative_variance
})

print(pca_summary)

fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.bar(pca_summary['PC'], pca_summary['Explained Variance Ratio'], color='skyblue', label='Individual Variance')
ax1.set_ylabel('Explained Variance Ratio')
ax1.set_xlabel('Principal Components')
ax1.set_title('PCA Explained Variance')

ax2 = ax1.twinx()
ax2.plot(pca_summary['PC'], pca_summary['Cumulative Variance'], color='red', marker='o', label='Cumulative Variance')
ax2.set_ylabel('Cumulative Explained Variance')
ax2.set_ylim(0, 1.1)

plt.show()
```
