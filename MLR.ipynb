Data Splitting and Preprocessing

```python
X=df.drop('Price',axis=1)
y=df['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

### Model building Linear regression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train_scaled,y_train)
```

```python
lr.coef_
```

### Model Evaluation

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
y_pred=lr.predict(X_test_scaled)
mse=mean_squared_error(y_test,y_pred)
print(mse)
```

```python
r2=r2_score(y_test,y_pred)
print(r2)
```

```python
mae=mean_absolute_error(y_test,y_pred)
print(mae)
```

### Ridge Regression

```python
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np
ridge = Ridge(alpha=1.0) 
ridge.fit(X_train_scaled, y_train)
ridge_pred = ridge.predict(X_test_scaled)

lasso = Lasso(alpha=1.0)
lasso.fit(X_train_scaled, y_train)
lasso_pred = lasso.predict(X_test_scaled)

def evaluate(y_true, y_pred, name):
    print(f"--- {name} Results ---")
    print(f"R2 Score: {r2_score(y_true, y_pred):.4f}")
    print(f"MAE: {mean_absolute_error(y_true, y_pred):.2f}")
    print(f"RMSE: {np.sqrt(mean_squared_error(y_true, y_pred)):.2f}\n")

evaluate(y_test, ridge_pred, "Ridge")
evaluate(y_test, lasso_pred, "Lasso")
```

## Coefficents of linear,Ridge,Lasso

```python
import pandas as pd

data = {
    'Feature': X.columns,
    'Linear_Regression': lr.coef_,
    'Ridge': ridge.coef_,
    'Lasso': lasso.coef_
}

df_coef = pd.DataFrame(data)

df_coef.set_index('Feature', inplace=True)

df_coef['Abs_Diff_LR_Lasso'] = (df_coef['Linear_Regression'] - df_coef['Lasso']).abs()

df_coef.sort_values(by='Linear_Regression', ascending=False)
```

## interpretation of your model’s coefficients

Age_08_04 (-2346.24): This is your most dominant feature. The large negative coefficient confirms that for every unit increase in age, the price drops significantly. In this model, age is roughly twice as influential as the next strongest feature.

Weight (+1014.59): This is the strongest positive predictor. Heavier cars in this dataset command a much higher price, likely because weight acts as a proxy for larger vehicle classes, better build quality, or higher trim levels.

KM (-608.82): Mileage has a notable negative impact on value. As kilometers increase, the price decreases, though the model currently views the "passage of time" (Age) as a much harsher penalty than the "distance driven" (KM).

Fuel_Type_Petrol (+376.67): Compared to the baseline (CNG), Petrol engines provide a solid positive bump to the predicted price.

HP (+371.32): Horsepower shows a clear positive correlation. More powerful engines lead to higher valuations, which aligns with standard market expectations for performance vehicles.

Fuel_Type_Diesel (+207.66): Being a Diesel car also adds value relative to CNG, though in this specific linear combination, it provides less of a "bonus" than Petrol.

Automatic (+50.56): This feature has a very low positive impact. While automatic cars are slightly more expensive, the model doesn't consider the transmission type a major driver of the final price compared to age or weight.

Doors (-29.20): This has a negligible negative effect. The number of doors is almost "noise" in this model, meaning it doesn't significantly help or hurt the price prediction once other factors are accounted for.

cc (-11.07): This is the weakest feature. The tiny negative coefficient suggests that engine capacity (cc) has almost no unique predictive power here, likely because its influence is already being "captured" by the Weight and HP features.
