Redundancy Check: The high correlation between RI and Ca suggests that a model might only need one of these to make accurate predictions, as they provide very similar information.

### Standerdization

```python
from sklearn.preprocessing import StandardScaler

X = df.drop('Type', axis=1)
y = df['Type']

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
print("Scaling Complete. Feature means are now ~0 and Std Dev is 1.")
```

### Train-Test Split

```python
from sklearn.model_selection import train_test_split

# Split the data
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

print(f"Training set size: {X_train.shape[0]} samples")
print(f"Testing set size: {X_test.shape[0]} samples")
```

### 3. Random Forest Implementation & Training

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

rf_model.fit(X_train, y_train)

print("Random Forest Model Training Complete.")
```

### 4. Model Evaluation

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

y_pred = rf_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)

print(f"Overall Model Accuracy: {accuracy:.2%}")
print("-" * 30)
print("Detailed Classification Report:")
print(classification_report(y_test, y_pred))

plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, cmap='Blues', fmt='g')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()
```

Implementation: Bagging (Random Forest)

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Initialize and train Bagging model (Random Forest)
bagging_model = RandomForestClassifier(n_estimators=100, random_state=42)
bagging_model.fit(X_train, y_train)

# Evaluate
y_pred_bag = bagging_model.predict(X_test)
bag_acc = accuracy_score(y_test, y_pred_bag)

print(f"Bagging (Random Forest) Accuracy: {bag_acc:.4f}")
```

2. Implementation: Boosting (Gradient Boosting)

```python
from sklearn.ensemble import GradientBoostingClassifier

# Initialize and train Boosting model
boosting_model = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
boosting_model.fit(X_train, y_train)

# Evaluate
y_pred_boost = boosting_model.predict(X_test)
boost_acc = accuracy_score(y_test, y_pred_boost)

print(f"Boosting (Gradient Boosting) Accuracy: {boost_acc:.4f}")
```

3. Comparison of Results

```python
import pandas as pd

comparison_metrics = {
    "Metric": ["Accuracy", "F1-Score (Macro)"],
    "Bagging (Random Forest)": [
        accuracy_score(y_test, y_pred_bag),
        classification_report(y_test, y_pred_bag, output_dict=True)['macro avg']['f1-score']
    ],
    "Boosting (Gradient Boosting)": [
        accuracy_score(y_test, y_pred_boost),
        classification_report(y_test, y_pred_boost, output_dict=True)['macro avg']['f1-score']
    ]
}

comparison_df = pd.DataFrame(comparison_metrics)
print(comparison_df)

comparison_df.set_index("Metric").plot(kind='bar', figsize=(10, 6), color=['skyblue', 'salmon'])
plt.title("Bagging vs Boosting Performance")
plt.ylabel("Score")
plt.xticks(rotation=0)
plt.legend(loc='lower right')
plt.show()
```

## Interview questions

# 2.Explain Bagging and Boosting methods. How is it different from each other.


1. Bagging vs. Boosting
Both Bagging (Bootstrap Aggregating) and Boosting are ensemble learning techniques that combine multiple weak models (usually decision trees) to create a single strong model.

Bagging (Bootstrap Aggregating)
How it works: It trains multiple base models in parallel on different random subsets of the training data, sampled with replacement (bootstrapping). The final prediction is made by averaging the outputs (for regression) or taking a majority vote (for classification).

Primary Goal: To reduce variance (overfitting).

Example: Random Forest.

Boosting
How it works: It trains base models sequentially. Each new model focuses on correcting the errors made by the previous models by assigning higher weights to misclassified or high-residual data points.

Primary Goal: To reduce bias (underfitting).

Example: XGBoost, AdaBoost, LightGBM.

Key Differences
Training Process: Bagging trains models in parallel independently; Boosting trains models sequentially where each model depends on the previous one.

Data Weighting: Bagging treats all data points equally; Boosting dynamically changes data weights based on previous errors.

Final Combination: Bagging uses simple voting or averaging; Boosting uses a weighted average based on each model's accuracy.

# 2.Explain how to handle imbalance in the data.

2. Handling Imbalanced Data
When one class severely outnumbers the other, machine learning models tend to favor the majority class. Here are the most effective ways to handle it:

Resampling Techniques
Oversampling (SMOTE): Synthetic Minority Over-sampling Technique creates synthetic, realistic examples of the minority class rather than just duplicating them.

Undersampling: Randomly removes samples from the majority class to match the minority class size (use only when you have abundant data).

Algorithmic Approaches
Class Weights (class_weight='balanced'): Most algorithms (like Logistic Regression, SVM, and Random Forest) allow you to penalize mistakes on the minority class more heavily during training.

Tree-Based Ensembles: Use boosting algorithms like XGBoost or LightGBM, which inherently handle imbalance better by focusing sequentially on hard-to-classify points.

Evaluation Metrics (Crucial)
Never use Accuracy: A model can have 99% accuracy by simply guessing the majority class every time.

Use Alternative Metrics: Focus on Precision, Recall, F1-Score, and the Area Under the Precision-Recall Curve (PR-AUC).
