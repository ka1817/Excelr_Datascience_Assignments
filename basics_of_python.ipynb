## Exercise 1: Prime Numbers

```python
def is_prime(number):
    """
    Checks if a given natural number is a prime number.
    A prime number is greater than 1 and has no positive divisors other than 1 and itself.
    """
    if number <= 1:
        return False
    
    if number == 2:
        return True
    
    if number % 2 == 0:
        return False
    for i in range(3, int(number**0.5) + 1, 2):
        if number % i == 0:
            return False  
            
    return True  

test_num = 29
if is_prime(test_num):
    print(f"Exercise 1 Result: {test_num} is a prime number.")
else:
    print(f"Exercise 1 Result: {test_num} is NOT a prime number.")
```

## Exercise 2: Product of Random Numbers

```python
import random

def random_product_quiz():
    """
    Generates two random integers, asks the user for their product,
    and validates if the user's answer is correct.
    """
    num1 = random.randint(1, 20)
    num2 = random.randint(1, 20)
    correct_product = num1 * num2
    
    print(f"--- Product Quiz ---")
    print(f"What is the product of {num1} and {num2}?")
    
    try:
        user_answer = int(input("Your answer: "))
        
        if user_answer == correct_product:
            print("Excellent! Your answer is correct. 🎉")
        else:
            print(f"Not quite. The correct answer was {correct_product}.")
    except ValueError:
        print("Invalid input! Please enter a valid integer next time.")

# Run the quiz function
random_product_quiz()
```

## Hospital Billing system

```python
# --- Hospital Billing System ---

ROOM_CHARGE_PER_DAY = 2000
DOCTOR_FEE = 1500
LAB_TEST_COST_PER_TEST = 300
DISCOUNT_THRESHOLD = 10000
DISCOUNT_RATE = 0.10  # 10%

# 2. Patient's specific data
days_stayed = 4
num_lab_tests = 3
medicine_charges = 2400

# 3. Calculate individual costs
total_room_charges = days_stayed * ROOM_CHARGE_PER_DAY
total_lab_charges = num_lab_tests * LAB_TEST_COST_PER_TEST

# 4. Calculate total bill before discount
total_bill_before_discount = (
    total_room_charges + DOCTOR_FEE + total_lab_charges + medicine_charges
)

# 5. Apply discount if eligible
discount_amount = 0.0
if total_bill_before_discount > DISCOUNT_THRESHOLD:
    discount_amount = total_bill_before_discount * DISCOUNT_RATE

# 6. Calculate final bill
final_bill = total_bill_before_discount - discount_amount

print("=" * 35)
print("       HOSPITAL BILL RECEIPT       ")
print("=" * 35)
print(f"Room Charges ({days_stayed} days)   : ₹{total_room_charges:,}")
print(f"Doctor Consultation Fee : ₹{DOCTOR_FEE:,}")
print(f"Lab Test Charges ({num_lab_tests} tests): ₹{total_lab_charges:,}")
print(f"Medicine Charges        : ₹{medicine_charges:,}")
print("-" * 35)
print(f"Total Before Discount   : ₹{total_bill_before_discount:,}")
print(f"Discount Applied (10%)  : ₹{discount_amount:,}")
print("-" * 35)
print(f"FINAL BILL TO PAY       : ₹{final_bill:,}")
print("=" * 35)
```

## Exercise 4: Word Counter

```python
def count_words(text):
    """
    Counts the frequency of words in a given string.
    Note: To exactly match the assignment sample output, 
    words retain their existing case and attached punctuation (e.g., 'text.').
    """
    words = text.split()
    
    word_counts = {}
    
    for word in words:
        if word in word_counts:
            word_counts[word] += 1
        else:
            word_counts[word] = 1
            
    return word_counts

input_text = "This is a sample text. This text will be used to demonstrate the word counter."
results = count_words(input_text)

print("--- Word Counts ---")
for word, count in results.items():
    print(f"'{word}': {count}")
```

## Exercise 5: Check for Palindrome

```python
import string

def is_palindrome(input_string):
    """
    Returns True if the string is a palindrome, and False otherwise.
    Ignores spaces, punctuation, and capitalization.
    """
    clean_string = input_string.lower()
    
    clean_string = clean_string.replace(" ", "")
    
    clean_string = clean_string.translate(str.maketrans('', '', string.punctuation))
    
    return clean_string == clean_string[::-1]

test_case_1 = "racecar"
test_case_2 = "A man, a plan, a canal: Panama" 
print(f"Is '{test_case_1}' a palindrome?: {is_palindrome(test_case_1)}") 
print(f"Is '{test_case_2}' a palindrome?: {is_palindrome(test_case_2)}")
```
