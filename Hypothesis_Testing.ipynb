### Assignment 1: Estimation And Confidence Intervals

Task a: Build 99% Confidence Interval Using Sample Standard Deviation
Rationale for Using the t-Distribution

We must use the Student's $t$-distribution for this task because:  

1) The population standard deviation is unknown. We only have access to the sample data to calculate a sample standard deviation ($s$).  

2) The sample size is small ($n = 15$, which is less than 30).  When the population is assumed to be normally distributed but is unknown, the test statistic follows a $t$-distribution with $n - 1$ degrees of freedom.

When the population is assumed to be normally distributed but standard deviation is unknown, the test statistic follows a $t$-distribution with $n - 1$ degrees of freedom.

```python
import numpy as np
import scipy.stats as stats

# 1. Input the durability data (in millions of characters)
data = np.array(
    [
        1.13,
        1.55,
        1.43,
        0.92,
        1.25,
        1.36,
        1.32,
        0.85,
        1.07,
        1.48,
        1.20,
        1.33,
        1.18,
        1.22,
        1.29,
    ]
)

# 2. Calculate sample statistics
n = len(data)
sample_mean = np.mean(data)
# ddof=1 ensures we get the sample standard deviation (divided by n-1)
sample_std = np.std(data, ddof=1)
degrees_of_freedom = n - 1
confidence_level = 0.99

# 3. Calculate the critical t-value
t_critical = stats.t.ppf((1 + confidence_level) / 2, df=degrees_of_freedom)

# 4. Calculate Margin of Error
margin_of_error_t = t_critical * (sample_std / np.sqrt(n))

# 5. Calculate Confidence Interval
ci_lower_t = sample_mean - margin_of_error_t
ci_upper_t = sample_mean + margin_of_error_t

print("--- Task a: Unknown Population Standard Deviation (t-distribution) ---")
print(f"Sample Mean: {sample_mean:.4f} million characters")
print(f"Sample Standard Deviation: {sample_std:.4f} million characters")
print(f"Critical t-value: {t_critical:.4f}")
print(
    f"99% Confidence Interval: ({ci_lower_t:.4f}, {ci_upper_t:.4f}) million characters\n"
)
```

### Task b: Build 99% Confidence Interval Using Known Population Standard Deviation

When the population standard deviation is known, we utilize the standard normal distribution ($z$-distribution) instead of the $t$-distribution, regardless of how small the sample size is

```python
# 1. Given population parameters
pop_std = 0.2  # Known population standard deviation
confidence_level = 0.99

# 2. Calculate the critical z-value
z_critical = stats.norm.ppf((1 + confidence_level) / 2)

# 3. Calculate Margin of Error
margin_of_error_z = z_critical * (pop_std / np.sqrt(n))

# 4. Calculate Confidence Interval
ci_lower_z = sample_mean - margin_of_error_z
ci_upper_z = sample_mean + margin_of_error_z

print(
    "--- Task b: Known Population Standard Deviation (z-distribution) ---"
)
print(f"Known Population Standard Deviation: {pop_std} million characters")
print(f"Critical z-value: {z_critical:.4f}")
print(
    f"99% Confidence Interval: ({ci_lower_z:.4f}, {ci_upper_z:.4f}) million characters"
)
```

## Assignment 2: Hypothesis Testing

#### Step 1: State the Hypotheses Statement

We are testing whether the actual weekly operating costs are higher than the theoretical model predicts. This requires a one-tailed (right-tailed) test.  

Null Hypothesis ($H_0$): $\mu \le \mu_{\text{theoretical}}$ (The true mean weekly operating cost is less than or equal to the cost model's expected value)

Alternative Hypothesis ($H_1$): $\mu > \mu_{\text{theoretical}}$ (The true mean weekly operating cost is higher than the model suggests). 

```python
import numpy as np
import scipy.stats as stats

print("--- Hypothesis Testing Process ---\n")

# 1. Theoretical Mean Calculation (μ)
# Model: W = 1000 + 5X, given X = 600 units
X_units = 600
mu_theoretical = 1000 + (5 * X_units)
print(f"Step 1 & 2 Parameters:")
print(f"Theoretical Mean Weekly Cost (μ): Rs. {mu_theoretical}")

# Given Sample Data
n = 25  # Sample size
sample_mean = 3050  # x-bar
# Population standard deviation for the cost equation: σ = 5 * 25 units
sigma_population = 5 * 25

print(f"Sample Mean (x_bar): Rs. {sample_mean}")
print(f"Population Standard Deviation (σ): {sigma_population}")
print(f"Sample Size (n): {n}")

# 2. Calculate the Test Statistic (z)
# Formula: z = (x_bar - μ) / (σ / sqrt(n))
standard_error = sigma_population / np.sqrt(n)
z_statistic = (sample_mean - mu_theoretical) / standard_error
print(f"\nCalculated z-statistic: {z_statistic:.4f}")

# 3. Determine Critical Value and p-value
alpha = 0.05
# Right-tailed critical z value
z_critical = stats.norm.ppf(1 - alpha)
p_value = 1 - stats.norm.cdf(z_statistic)

print(f"\nStep 3: Probability and Thresholds:")
print(f"Significance Level (α): {alpha}")
print(f"Critical z-value: {z_critical:.4f}")
print(f"Calculated p-value: {p_value:.6e}")

# 4. Make a Decision
print(f"\nStep 4: Decision:")
if z_statistic > z_critical:
    decision = "Reject the Null Hypothesis (H0)"
else:
    decision = "Fail to Reject the Null Hypothesis (H0)"
print(f"Decision: {decision}")

# 5. Conclusion
print(f"\nStep 5: Conclusion:")
if z_statistic > z_critical:
    print(
        "Conclusion: There is strong evidence to support the restaurant owners' claim "
        "that the weekly operating costs are significantly higher than what the theoretical model suggests."
    )
else:
    print(
        "Conclusion: There is insufficient evidence to support the restaurant owners' claim. "
        "The observed operating costs are not statistically higher than the model predictions."
    )
```
