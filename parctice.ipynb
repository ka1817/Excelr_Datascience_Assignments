```python
class New():
    def __init__(self,name,roll):
        self.name=name
        self.roll=roll 
    def into(self):
        print(f"My name is {self.name} and roll no id {self.roll}")
obj=New('pranav',18)

obj.into()
```

```python
class calculator:
    def __init__(self,name):
        self.name=name 
    def add(self,a,b):
        return a+b 
    def sub(self,a,b):
        return a-b 
    def mul(self,a,b):
        return a*b 
    def div(self,a,b):
        return a/b 
cal=calculator("calc")
cal.mul(4,5)

```

```python
class latest:
    def __init__(self,name,age):
        self.name=name 
        self.age=age 
    def details(self):
        print(f"The name of the student is {self.name} and the age of the student is {self.age}")
class New1(latest):
    def __init__(self,name,age,roll):
        super().__init__(name,age)
        self.roll=roll
    def pred(self):
        print(f"The age of student is {self.age} and roll is {self.roll}")

c=New1("pranav",23,61)
c.details()
c.pred()
```
