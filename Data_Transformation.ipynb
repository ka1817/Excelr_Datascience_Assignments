```python
import math
import seaborn as sns
import matplotlib.pyplot as plt

cat_features = ['workclass', 'education', 'marital_status', 'occupation', 
                'relationship', 'race', 'sex', 'native_country', 'income']

n_cols = 3
n_rows = math.ceil(len(cat_features) / n_cols)

sns.set_theme(style="whitegrid")
fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(18, 15))
fig.suptitle('Count Plots for Categorical & Ordinal Features', fontsize=22)

axes = axes.flatten()

edu_order = ['Preschool', '1st-4th', '5th-6th', '7th-8th', '9th', '10th', '11th', '12th', 
             'HS-grad', 'Some-college', 'Assoc-voc', 'Assoc-acdm', 'Bachelors', 
             'Masters', 'Prof-school', 'Doctorate']

for i, col in enumerate(cat_features):
    plot_order = edu_order if col == 'education' else df[col].value_counts().index
    
    sns.countplot(data=df, x=col, ax=axes[i], palette='viridis', hue=col, 
                  legend=False, order=plot_order)
    
    axes[i].set_title(f'Count of {col}', fontsize=14)
    axes[i].set_xlabel('') 
    axes[i].set_ylabel('Count')
    
    axes[i].tick_params(axis='x', rotation=45)

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

## 1.Ordinal Features 

These are categorical variables where the categories have a clear "rank" or "hierarchy.

"Education": This is the most distinct ordinal feature. There is a logical progression from Preschool $\rightarrow$ HS-grad $\rightarrow$ Bachelors $\rightarrow$ Masters $\rightarrow$ Doctorate.

"Income (Target)": Since this represents a threshold ($<=50K$ vs $>50K$), it is technically ordinal (or binary-ordinal) because one category represent a higher value/status than the other.



## 2.Nominal Features

These features have no mathematical order; one category is not "higher" or "better" than another.

Workclass: (e.g., Private, Self-emp, State-gov) — Different sectors of employment, no intrinsic ranking.

Marital Status: (e.g., Married, Divorced, Never-married) — Social states with no mathematical order.

Occupation: (e.g., Exec-managerial, Craft-repair, Sales) — Different types of jobs.

Relationship: (e.g., Husband, Own-child, Unmarried) — Roles within a household.

Race: (e.g., White, Black, Asian-Pac-Islander) — Demographic categories.

Sex: (Male, Female) — Biological categories.

Native Country: Geographic origins.

Handling Missing Values

```python
import numpy as np
df.replace('?', np.nan, inplace=True)

num_cols = df.select_dtypes(include=['int64', 'float64']).columns
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

cat_cols = df.select_dtypes(include=['object']).columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

print("Missing values handled.")
```

Applying both Standard Scaling and Min-Max Scaling to numerical features.

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

numerical_features = ['age', 'fnlwgt', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week']

# a. Standard Scaling (Mean=0, Std=1)
std_scaler = StandardScaler()
df_standard = df.copy()
df_standard[numerical_features] = std_scaler.fit_transform(df[numerical_features])

# b. Min-Max Scaling (Range 0 to 1)
mm_scaler = MinMaxScaler()
df_minmax = df.copy()
df_minmax[numerical_features] = mm_scaler.fit_transform(df[numerical_features])

print("Scaling applied to separate dataframes.")
```

Note on Scenarios:

Standard Scaling is preferred for algorithms that assume a Gaussian distribution (e.g., Logistic Regression, SVM) as it is less sensitive to outliers.

Min-Max Scaling is preferred when the distribution is not Gaussian or for algorithms that require a specific bounded range (e.g., Neural Networks or K-Nearest Neighbors).

Encoding Techniques

Applying One-Hot Encoding for features with $< 5$ categories and Label Encoding for those with $\ge 5$.

```python
from sklearn.preprocessing import LabelEncoder

cat_cols = df.select_dtypes('object').columns
ohe_cols = [col for col in cat_cols if df[col].nunique() < 5]
label_cols = [col for col in cat_cols if df[col].nunique() >= 5]

df = pd.get_dummies(df, columns=ohe_cols, drop_first=True)

le = LabelEncoder()
for col in label_cols:
    df[col] = le.fit_transform(df[col])

print(f"One-Hot Encoded: {ohe_cols}")
print(f"Label Encoded: {label_cols}")
```

Pros/Cons:

One-Hot: Prevents the model from assuming a false numerical order but increases dimensionality ("Curse of Dimensionality").

Label: Keeps the dataset compact but can mislead models into thinking "4" is "greater than" "1" in a non-ordinal feature.

Feature Engineering & Log Transformation

Creating two new features and applying log transformation to skewed data.

```python
df['capital_diff'] = df['capital_gain'] - df['capital_loss']

df['age_bins'] = pd.cut(df['age'], bins=[0, 25, 45, 65, 100], labels=[1, 2, 3, 4])

df['capital_gain_log'] = np.log1p(df['capital_gain'])

print("Feature engineering and transformations complete.")
```

Rationale: > * Capital_Diff: Combines two related financial indicators into a single "net" value.

Age_Bins: Reduces noise by grouping individuals into life stages (young, prime, senior).

Log Transformation: Applied to capital_gain because it is highly right-skewed; this helps normalize the distribution for better model convergence.
