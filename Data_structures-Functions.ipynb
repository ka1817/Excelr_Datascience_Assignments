### Section A: Data Structures & Control Structures

#### Question 1: List Operations
Create a list of 5 integers. Perform and print the result of various list operations.  

```python
# Create a list of 5 integers
my_list = [40, 10, 30, 20, 50]
print("Original List:", my_list)

# 1. append: Adds an element to the end
my_list.append(60)
print("After append(60):", my_list)

# 2. extend: Appends elements from an iterable to the end
my_list.extend([70, 80])
print("After extend([70, 80]):", my_list)

# 3. insert: Inserts an item at a given index
my_list.insert(2, 25)  # Inserts 25 at index 2
print("After insert(2, 25):", my_list)

# 4. remove: Removes the first item with the specified value
my_list.remove(10)
print("After remove(10):", my_list)

# 5. pop: Removes and returns the item at the given index (default last)
popped_item = my_list.pop()
print(f"After pop(): {my_list} (Popped item: {popped_item})")

# 6. index: Returns the index of the first matched value
idx = my_list.index(30)
print(f"Index of 30: {idx}")

# 7. count: Returns the number of times a value appears
cnt = my_list.count(20)
print(f"Count of 20: {cnt}")

# 8. sort: Sorts the list in ascending order
my_list.sort()
print("After sort():", my_list)

# 9. reverse: Reverses the elements of the list in place
my_list.reverse()
print("After reverse():", my_list)

# 10. clear: Removes all items from the list
my_list.clear()
print("After clear():", my_list)
```

#### Question 2: Tuples
Create a tuple that stores 3 student names. Try changing the second name in the tuple.

```python
students = ("Amit", "Rahul", "Priya")
print("Original Tuple:", students)

try:
    students[1] = "Rohit"
except TypeError as e:
    print(f"\n[ERROR OCCURRED]: TypeError: {e}")

print("\n--- Explanation ---")
print(
    "Tuples are IMMUTABLE data structures in Python. Once a tuple is created, "
    "its elements cannot be modified, added, or removed. Attempting to item-assign "
    "raises a TypeError."
)
```

#### Question 5: Conditional Script (Voting Eligibility)
Write a script that checks if a person is eligible to vote (age $\ge$ 18)

```python
# Take age as a variable
age = 20

# Check eligibility
if age >= 18:
    print(f"Age: {age} -> Eligible to vote!")
else:
    print(f"Age: {age} -> Not eligible to vote yet.")
```

#### Question 6: Grade Evaluation
Given a 'marks' variable, print the grade.

```python
# Marks variable
marks = 82

# Grade control flow evaluation
if marks >= 90:
    grade = "A"
elif marks >= 75:
    grade = "B"
elif marks >= 50:
    grade = "C"
else:
    grade = "Fail"

print(f"Marks: {marks} -> Grade: {grade}")
```

#### Question 7: Nested Conditions
Given a number, check if it's positive, and if it is also even. If not positive, print if it's zero or negative.

```python
# Input number
num = 14

if num > 0:
    if num % 2 == 0:
        print(f"{num} is Positive and Even.")
    else:
        print(f"{num} is Positive but Odd.")
elif num == 0:
    print(f"{num} is Zero.")
else:
    print(f"{num} is Negative.")
```

### Section C: Exploring Pandas

#### Question 11: Dataframe Exploration & Manipulation

Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then execute downstream operations.

```python
import numpy as np
import pandas as pd

raw_data = {
    "Name": [
        "Amit",
        "Rahul",
        "Priya",
        "Amit",
        "Sonia",
        "Vijay",
        "Neha",
        "Raj",
        "Anjali",
        "Vikram",
    ],
    "Age": [23, 28, 25, 23, 31, 42, 27, 35, 29, 33],
    "City": [
        "Delhi",
        "Mumbai",
        "Bangalore",
        "Delhi",
        "Chennai",
        "Delhi",
        "Mumbai",
        "Pune",
        "Bangalore",
        "Pune",
    ],
    "Salary": [
        50000,
        62000,
        np.nan,
        50000,
        71000,
        85000,
        np.nan,
        90000,
        58000,
        65000,
    ],  
}

df = pd.DataFrame(raw_data)
print("--- Step 1: Initial Manually Created DataFrame ---")
print(df)
print("-" * 50)

print("\n--- Step 2a: .info() Summary ---")
df.info()

print("\n--- Step 2b: .describe() Statistical Breakdown ---")
print(df.describe())
print("-" * 50)

name_city_df = df[["Name", "City"]]
print("\n--- Step 3: Selected 'Name' and 'City' Columns ---")
print(name_city_df.head())
print("-" * 50)

df_dropped_city = df.drop(columns=["City"])
print("\n--- Step 4: DataFrame After Dropping 'City' Column ---")
print(df_dropped_city.head())
print("-" * 50)

mean_salary = df["Salary"].mean()
df["Salary"] = df["Salary"].fillna(mean_salary)
print(f"\n--- Step 5: Filled Missing Salaries with Mean (₹{mean_salary:.2f}) ---")
print(df)
print("-" * 50)

df_cleaned = df.drop_duplicates()
print("\n--- Step 6: Final Cleaned DataFrame (Duplicates Removed) ---")
print(df_cleaned)
```
